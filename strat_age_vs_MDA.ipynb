```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
xl = pd.ExcelFile("C:/Users/culle/Documents/Data/Tornillo_MDAs.xlsx")
df = xl.parse("Sheet1")
df.set_index('section', inplace=True)

df[:10]
```

```python
df2 = df.loc['Dawson Creek']

df2
```

```python
#mdaList = ['Tie-Point', 'YDZA', 'YC3+', 'YSG']

mdaList = ['DSA', 'YC3+', 'YSG']

mdaList
```

```python
#X-Y values for the plot
x = []
x_err = []
y = []

for i in range(len(mdaList)): 
    x.append(df2.loc[df2['mda_type']==mdaList[i]]['mda'])
    x_err.append(df2.loc[df2['mda_type']==mdaList[i]]['mda_err'])
    y.append(df2.loc[df2['mda_type']==mdaList[i]]['strat_m'])

x
```

```python
fig, ax = plt.subplots(figsize=(7.5, 10))

#ax.plot((55.5,55.5), (350, 500), color='black', linestyle='dashed')
ax.plot((66,66), (140, 220), color='black', linestyle='dashed')

#Dawson Creek Fossils
ax.fill((71.7, 71.7, 71.5, 71.5), (0, 5, 5, 0), color='#fed98e', alpha=1)
ax.fill((69.7, 69.7, 69.2, 69.2), (5, 100, 100, 5), color='#fed98e', alpha=1)
ax.fill((66.5, 66.5, 65.9, 65.9), (100, 195, 195, 100), color='#fed98e', alpha=1)
ax.fill((63.5, 63.5, 63, 63), (195, 247, 247, 195), color='#fed98e', alpha=1)

ax.plot((66, 70.6), (6, 6), color='#993404')
ax.plot((66, 70.6), (100, 100), color='#993404')
ax.plot((66, 70.6), (150, 150), color='#993404')
ax.plot((66, 70.6), (155, 155), color='#993404')
ax.plot((66, 70.6), (158, 158), color='#993404')
ax.plot((66, 70.6), (170, 170), color='#993404')
ax.plot((60.9, 63.8), (197, 197), color='#993404')
ax.plot((60.9, 63.8), (204, 204), color='#993404')

#Tornillo Mammal Fossils
#ax.plot((66, 70.6), (145, 145), color='#993404')
#ax.plot((56.2, 63.8), (295, 295), color='#993404')
#ax.plot((56.2, 60.9), (325, 325), color='#993404')
#ax.plot((56.2, 60.9), (355, 355), color='#993404')
#ax.plot((56.2, 60.9), (377, 377), color='#993404')
#ax.plot((56.2, 60.9), (400, 400), color='#993404')
#ax.plot((50.5, 54.9), (435, 435), color='#993404')
#ax.plot((50.5, 54.9), (482, 482), color='#993404')
#ax.plot((46.2, 50.5), (545, 545), color='#993404')
#ax.plot((46.2, 50.5), (560, 560), color='#993404')

for i in range(len(x)):
    colors = plt.cm.YlOrRd_r(np.linspace(0.4, 1, len(mdaList)))
    #markers = ['P', 'd', '^', 'v']
    markers = ['d', '^', 'v']
    labels = mdaList
    ax.errorbar((x[i]), y[i], xerr=(x_err[i]), yerr=0, linestyle="", color='black', capsize=0)
    ax.plot((x[i]), (y[i]), marker=(markers[i]), markersize=8, color=(colors[i]), markeredgecolor='black', alpha=1, linestyle="", label=(labels[i]))

ax.set_xlim(30, 85)
#ax.set_ylim(75, 50)

ax.set_title ("Dawson Creek Maximum Depositional Ages")
ax.set_xlabel('Maximum Depositional Age (Ma)')
ax.set_ylabel('Height in Stratigraphic Column (m)')
ax.legend()
```
