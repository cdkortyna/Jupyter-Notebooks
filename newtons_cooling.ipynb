```python
import math
from matplotlib import pyplot as plt

def find_cooling_constants(t1, t2, tdiff, ta):
    C = t1 - ta
    k = math.log((t2 - ta) / C) / -tdiff
    return C, k

def calculate_cooling(t, C, k, ta):
    T = C * math.exp(-k * t) + ta
    return T
```

```python
t1 = 65
t2 = 30
tdiff = 8
ta = 12

C, k = find_cooling_constants(t1, t2, tdiff, ta)
#print(C, k)

times = []
T = []
for t in range(0, 24):
    Ti = calculate_cooling(t, C, k, ta)
    times.append(t)
    T.append(Ti)

plt.figure(figsize=(12,8))
plt.plot(times, T, marker = 'o')
plt.plot((0, 24), (12, 12), linestyle='dashed')
plt.xlabel('Hours Since Power Shutoff')
plt.ylabel('Temperature (F)')
plt.title("Modeled Rate of Heat Loss: Zach's Cottage")
plt.text(0, 13, f'Ambient Air Temperature = {ta} F')
plt.show()
print(f'Starting Temp: {t1} F\tTemp After 12 Hours: {T[12]:0.2f} F' )
```
