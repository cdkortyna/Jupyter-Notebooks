```python
fig, ax = plt.subplots(figsize=(12, 10))
ax.set(yscale='log', xscale='log')
sns.scatterplot(data=df, x='Lb (km)', y='A (km^2)', hue='Ss (Mt/yr)', palette = 'YlGnBu', alpha = 0.75, 
                edgecolor = 'black', hue_norm = clr.LogNorm(vmin=0.1, vmax = 1000), s=100)

sns.scatterplot(data=df2, x='Lb (km)', y='A (km^2)', color = 'orange', alpha = 0.75, 
                edgecolor = 'black', marker = 'd', s=150)

#sns.scatterplot(data=df[27:28], x='Lb (km)', y='A (km^2)', color = 'yellow', alpha = 0.75, 
#                edgecolor = 'black', marker = 'd', s=150)

ax.set_title("Modern Big Rivers")
ax.set_xlabel("Drainage Basin Length (km)")
ax.set_ylabel("Drainage Basin Area (km$^2$)")

legend = ax.legend(title = "Suspended Load (Mt/yr)")
legend._legend_box.align = "left"
```
